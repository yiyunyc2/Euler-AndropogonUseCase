# Modeling four taxonomies of the Andropogon Complex

Examples from: 

Franz, N. M., Chen, M., Kianmajd, P., Yu, S., Bowers, S., Weakley, A. S., & Lud√§scher, B. (2016). Names are not good enough: reasoning over taxonomic change in the Andropogon complex 1. Semantic Web, 7(6), 645-667.
https://content.iospress.com/articles/semantic-web/sw220



Screen Shot 2023-01-16 at 8.20.14 PM.png

Four taxonomies modeled using the PROV model: 

Campbell(1983), Campbell(2003), Weakley(2006), Weakley (2015)

```python
from prov.model import ProvDocument
```

```python
# Create a new provenance document
d1 = ProvDocument()  # d1 is now an empty provenance document

# Declaring namespaces for various prefixes used in the example
d1.add_namespace('eg', 'http://www.example.org/')
```

```python
# Declaring agents 
d1.agent('eg:campbell', {'prov:type': 'prov:Person', 'eg:name': "Christopher S Campbell"})
d1.agent('eg:weakley', {'prov:type': 'prov:Person', 'eg:name': "Alan S Weakley"})
```

```python
# Declaring taxonomies as entities
d1.entity('eg:Andropogon_1983', {'prov:type': "eg:taxonomy"})
d1.entity('eg:Andropogon_2003', {'prov:type': "eg:taxonomy"})
d1.entity('eg:Andropogon_2006', {'prov:type': "eg:taxonomy"})
d1.entity('eg:Andropogon_2015', {'prov:type': "eg:taxonomy"})
```

```python
# Declaring new species as activities
#1983
d1.activity('eg:A_virginicus_var_glaucus_drylandsVariant')
d1.activity('eg:A_virginicus_var_glaucus_wetlandsVariant')
d1.activity('eg:A_virginicus_var_virginicus_oldfieldVariant')
d1.activity('eg:A_virginicus_var_virginicus_smoothVariant')
d1.activity('eg:A_virginicus_var_virginicus_deceptiveVariant')
d1.activity('eg:A_glomeratus_var_glaucopsis')
d1.activity('eg:A_glomeratus_var_hirsutior')
d1.activity('eg:A_glomeratus_var_glomeratus')
d1.activity('eg:A_glomeratus_var_pumilus')

#2003
d1.activity('eg:A_virginicus_var_decipiens')

#2006
d1.activity('eg:A_capillipes_drylandsVariant')
d1.activity('eg:A_capillipes_wetlandsVariant')
d1.activity('eg:A_virginious_var_virginicus')
d1.activity('eg:A_glaucopsis')
d1.activity('eg:A_tenuispatheus')

#2015
d1.activity('eg:A_capillipes')
d1.activity('eg:A_dealbatus')
d1.activity('eg:A_virginicus_oldFieldVariant')
d1.activity('eg:A_virginicus_smoothVariant')
d1.activity('eg:A_hirsutior')
```

```python
#Adding attribute to relationships
d1.wasAttributedTo('eg:Andropogon_1983', 'eg:campbell')
d1.wasAttributedTo('eg:Andropogon_2003', 'eg:campbell')
d1.wasAttributedTo('eg:Andropogon_2006', 'eg:weakley')
d1.wasAttributedTo('eg:Andropogon_2015', 'eg:weakley')
```

```python
#Adding derive from relationsihps
d1.wasDerivedFrom('eg:Andropogon_2003','eg:Andropogon_1983')
d1.wasDerivedFrom('eg:Andropogon_2015','eg:Andropogon_2006')
```

```python
#Adding generate by relationships
d1.wasGeneratedBy('eg:Andropogon_1983','eg:A_virginicus_var_glaucus_drylandsVariant',)
d1.wasGeneratedBy('eg:Andropogon_1983','eg:A_virginicus_var_glaucus_wetlandsVariant')
d1.wasGeneratedBy('eg:Andropogon_1983','eg:A_virginicus_var_virginicus_oldfieldVariant')
d1.wasGeneratedBy('eg:Andropogon_1983','eg:A_virginicus_var_virginicus_smoothVariant')
d1.wasGeneratedBy('eg:Andropogon_1983','eg:A_virginicus_var_virginicus_deceptiveVariant')
d1.wasGeneratedBy('eg:Andropogon_1983','eg:A_glomeratus_var_glaucopsis')
d1.wasGeneratedBy('eg:Andropogon_1983','eg:A_glomeratus_var_hirsutior')
d1.wasGeneratedBy('eg:Andropogon_1983','eg:A_glomeratus_var_glomeratus')
d1.wasGeneratedBy('eg:Andropogon_1983','eg:A_glomeratus_var_pumilus')

#2003
d1.wasGeneratedBy('eg:Andropogon_2003','eg:A_virginicus_var_decipiens') #new
d1.wasGeneratedBy('eg:Andropogon_2003','eg:A_virginicus_var_glaucus_drylandsVariant',)
d1.wasGeneratedBy('eg:Andropogon_2003','eg:A_virginicus_var_glaucus_wetlandsVariant')
d1.wasGeneratedBy('eg:Andropogon_2003','eg:A_virginicus_var_virginicus_oldfieldVariant')
d1.wasGeneratedBy('eg:Andropogon_2003','eg:A_virginicus_var_virginicus_smoothVariant')
d1.wasGeneratedBy('eg:Andropogon_2003','eg:A_glomeratus_var_glaucopsis')
d1.wasGeneratedBy('eg:Andropogon_2003','eg:A_glomeratus_var_hirsutior')
d1.wasGeneratedBy('eg:Andropogon_2003','eg:A_glomeratus_var_glomeratus')
d1.wasGeneratedBy('eg:Andropogon_2003','eg:A_glomeratus_var_pumilus')

#2006
d1.wasGeneratedBy('eg:Andropogon_2006','eg:A_capillipes_drylandsVariant')#new
d1.wasGeneratedBy('eg:Andropogon_2006','eg:A_capillipes_wetlandsVariant') #new
d1.wasGeneratedBy('eg:Andropogon_2006','eg:A_virginious_var_virginicus') #new
d1.wasGeneratedBy('eg:Andropogon_2006','eg:A_glaucopsis') #new
d1.wasGeneratedBy('eg:Andropogon_2006','eg:A_tenuispatheus') #new
d1.wasGeneratedBy('eg:Andropogon_2006','eg:A_virginicus_var_decipiens')
d1.wasGeneratedBy('eg:Andropogon_2006','eg:A_glomeratus_var_hirsutior')
d1.wasGeneratedBy('eg:Andropogon_2006','eg:A_glomeratus_var_glomeratus')



#2015
d1.wasGeneratedBy('eg:Andropogon_2015','eg:A_capillipes')#new
d1.wasGeneratedBy('eg:Andropogon_2015','eg:A_dealbatus')#new
d1.wasGeneratedBy('eg:Andropogon_2015','eg:A_virginicus_oldFieldVariant')#new
d1.wasGeneratedBy('eg:Andropogon_2015','eg:A_virginicus_smoothVariant')#new
d1.wasGeneratedBy('eg:Andropogon_2015','eg:A_hirsutior')#new
d1.wasGeneratedBy('eg:Andropogon_2015','eg:A_virginicus_var_decipiens')
d1.wasGeneratedBy('eg:Andropogon_2015','eg:A_glaucopsis') 
d1.wasGeneratedBy('eg:Andropogon_2015','eg:A_glomeratus_var_glomeratus')
d1.wasGeneratedBy('eg:Andropogon_2015','eg:A_tenuispatheus')


```

```python
#Adding Used relationships --- new species discovered/split/lumped first in which taxonomy 
#1983
d1.used('eg:A_virginicus_var_glaucus_drylandsVariant','eg:Andropogon_1983')
d1.used('eg:A_virginicus_var_glaucus_wetlandsVariant','eg:Andropogon_1983')
d1.used('eg:A_virginicus_var_virginicus_oldfieldVariant','eg:Andropogon_1983')
d1.used('eg:A_virginicus_var_virginicus_smoothVariant','eg:Andropogon_1983')
d1.used('eg:A_virginicus_var_virginicus_deceptiveVariant','eg:Andropogon_1983')
d1.used('eg:A_glomeratus_var_glaucopsis','eg:Andropogon_1983')
d1.used('eg:A_glomeratus_var_hirsutior','eg:Andropogon_1983')
d1.used('eg:A_glomeratus_var_glomeratus','eg:Andropogon_1983')
d1.used('eg:A_glomeratus_var_pumilus','eg:Andropogon_1983')

#2003
d1.used('eg:A_virginicus_var_decipiens','eg:Andropogon_2003')

#2006
d1.used('eg:A_capillipes_drylandsVariant','eg:Andropogon_2006')
d1.used('eg:A_capillipes_wetlandsVariant','eg:Andropogon_2006')
d1.used('eg:A_virginious_var_virginicus','eg:Andropogon_2006')
d1.used('eg:A_glaucopsis','eg:Andropogon_2006')
d1.used('eg:A_tenuispatheus','eg:Andropogon_2006')

#2015
d1.used('eg:A_capillipes','eg:Andropogon_2015')
d1.used('eg:A_dealbatus','eg:Andropogon_2015')
d1.used('eg:A_virginicus_oldFieldVariant','eg:Andropogon_2015')
d1.used('eg:A_virginicus_smoothVariant','eg:Andropogon_2015')
d1.used('eg:A_hirsutior','eg:Andropogon_2015')
```

```python
#Adding associated with relationships 
#1983
d1.wasAssociatedWith('eg:A_virginicus_var_glaucus_drylandsVariant','eg:campbell')
d1.wasAssociatedWith('eg:A_virginicus_var_glaucus_wetlandsVariant','eg:campbell')
d1.wasAssociatedWith('eg:A_virginicus_var_virginicus_oldfieldVariant','eg:campbell')
d1.wasAssociatedWith('eg:A_virginicus_var_virginicus_smoothVariant','eg:campbell')
d1.wasAssociatedWith('eg:A_virginicus_var_virginicus_deceptiveVariant','eg:campbell')
d1.wasAssociatedWith('eg:A_glomeratus_var_glaucopsis','eg:campbell')
d1.wasAssociatedWith('eg:A_glomeratus_var_hirsutior','eg:campbell')
d1.wasAssociatedWith('eg:A_glomeratus_var_glomeratus','eg:campbell')
d1.wasAssociatedWith('eg:A_glomeratus_var_pumilus','eg:campbell')

#2003
d1.wasAssociatedWith('eg:A_virginicus_var_decipiens','eg:campbell')

#2006
d1.wasAssociatedWith('eg:A_capillipes_drylandsVariant','eg:weakley')
d1.wasAssociatedWith('eg:A_capillipes_wetlandsVariant','eg:weakley')
d1.wasAssociatedWith('eg:A_virginious_var_virginicus','eg:weakley')
d1.wasAssociatedWith('eg:A_glaucopsis','eg:weakley')
d1.wasAssociatedWith('eg:A_tenuispatheus','eg:weakley')

#2015
d1.wasAssociatedWith('eg:A_capillipes','eg:weakley')
d1.wasAssociatedWith('eg:A_dealbatus','eg:weakley')
d1.wasAssociatedWith('eg:A_virginicus_oldFieldVariant','eg:weakley')
d1.wasAssociatedWith('eg:A_virginicus_smoothVariant','eg:weakley')
d1.wasAssociatedWith('eg:A_hirsutior','eg:weakley')
```
